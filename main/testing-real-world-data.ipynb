```python
dataset_directory = '../dataset'
test_directory = dataset_directory + '/test'
```

```python
from keras.utils import image_dataset_from_directory

image_size = (256, 256)

train_dataset = image_dataset_from_directory(
    test_directory,
    label_mode = 'categorical',
    image_size = image_size,
    batch_size = 3,
    shuffle = True
)
```

```python
from keras.models import load_model

resnet50 = load_model('../models/res-net-50-model.keras')
vgg16 = load_model('../models/vgg-16-model.keras')
inceptionv3 = load_model('../models/inception-v3-model.keras')
mymodel = load_model('../models/my-model.keras')
```

```python
from utils.evaluate_model import evaluate_model
```

## ResNet50

```python
evaluate_model(resnet50, train_dataset)
```

## VGG16

```python
evaluate_model(vgg16, train_dataset)
```

```python
evaluate_model(inceptionv3, train_dataset)
```
